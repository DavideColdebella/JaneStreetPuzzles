```python
from typing import List
import numpy as np
```

```python
def naive(t: List[int]) -> int:
    """
    
    :param t: a list of 4 elements
    :returns: the number of steps to reduce t to the e
    """
    c = 1
    while sum(t)!=0:
        t = [
            abs(t[1] - t[0]), 
            abs(t[2] - t[1]), 
            abs(t[3] - t[2]), 
            abs(t[0] - t[3])
        ]
        c+=1
    return c
```

```python
# Case I <<<
MI = np.asarray(
    [[-2, 1, 0],
    [-1, -1, 1],
    [-1, 0, 1]]
)
MI_inv = np.linalg.inv(MI)
# Case III >>>
MIII = np.asarray(
    [[-1, 2, -1],
    [1, 1, -1],
    [0, 1, 0]]
)
MIII_inv = np.linalg.inv(MIII)
```

```python
def integrate(F, G, H):
    vals = []

    if not (F<G<H):
        # print(f"{F}, {G}, {H}, invalid hypothesis")
        return vals
    if (H<=G-F):
        # print(f"{F}, {G}, {H}, not case I nor III")
        return vals
        
    def is_ok(F, G, H):
        if F<=0 or G<=0 or H<=0:
            # print(f"{F}, {G}, {H}, negative or zero")
            return False
        if sum(abs(np.rint([F, G, H]) - [F, G, H]))>10**-8:
            # print(f"{F}, {G}, {H}, float")
            return False
        if not (F<G<H):
            # print(f"{F}, {G}, {H}, wrong order")
            return False
        return True
    
    def rint(F, G, H):
        F, G, H = np.rint([F, G, H])
        return F, G, H

    # case III & I
    if H>G+F:
        for M_inv in [MI_inv, MIII_inv]:
            arr = np.dot(M_inv, [F, G, H])
            if is_ok(*arr):
                vals.append(list(rint(*arr)))
                
    return vals
    
```

```python
res = []
ans = 0

stack = []
for i in range(0, 24, 1):
    for j in range(i+1, 24, 1):
        for k in range(j+1, 24, 1):
            stack.append([2**i, 2**j, 2**k, 1])

while stack:
    F, G, H, cnt = stack.pop()
    # print(F, G, H, cnt)
    for t in integrate(F, G, H):
        if H <= 10**7:
            stack.append(t + [cnt +1])
            if cnt+1>ans:
                ans = cnt+1
                res = [t]
            elif cnt+1==ans:
                res.append(t)
```
